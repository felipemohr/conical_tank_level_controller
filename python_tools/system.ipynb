```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('system1_data.csv', delimiter=',')

millis = data[:,0]
dist_raw = data[:,1]
dist_filtered = data[:,2]
water_height = data[:,3]
setpoint = data[:,4]
pid_val = data[:,5]

error = 100*setpoint - water_height
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(20,12))
plt.title("Resposta do Sistema Real", fontweight="bold")
plt.xlabel("Tempo (s)")
plt.ylabel("Altura de gua (cm)")
plt.plot(millis, water_height, linewidth=2.2)
plt.plot(millis, 100*setpoint)
plt.grid()
```

```python
plt.figure(figsize=(20,12))
plt.plot(millis, dist_raw)
plt.plot(millis, dist_filtered)
```
