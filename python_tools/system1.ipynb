```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('system2.csv', delimiter=',')

millis = data[:,0]
dist_raw = data[:,1]
dist_filtered = data[:,2]
water_height = data[:,3]
setpoint = data[:,4]
#pid_val = data[:,5]

```

```python
plt.figure(figsize=(20,12))
plt.title("Resposta")
plt.xlabel("Tempo (s)")
plt.ylabel("Altura de água (cm)")
plt.plot(millis, water_height)
plt.plot(millis, 100*setpoint)
```

```python
plt.figure(figsize=(20,12))
plt.plot(millis, dist_raw)
plt.plot(millis, dist_filtered)
```
