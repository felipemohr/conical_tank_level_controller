```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt('system3_data.csv', delimiter=',')

millis = data[:,0]
setpoint = data[:,1]
flow_raw = data[:,2]
flow_filtered = data[:,3]

error = setpoint - flow_filtered
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(20,12))
plt.title("Resposta do Sistema Real", fontweight="bold")
plt.xlabel("Tempo (s)")
plt.ylabel("Vazo (L/min)")
plt.plot(millis, flow_filtered, linewidth=2.2)
plt.plot(millis, setpoint)
plt.grid()
```

```python
plt.figure(figsize=(20,12))
plt.plot(millis, flow_raw)
plt.plot(millis, flow_filtered)
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(20,12))
plt.title("Erro do Sistema Real", fontweight="bold")
plt.xlabel("Tempo (s)")
plt.ylabel("Erro (L/min)")
plt.plot(millis, error, linewidth=2.2, color='r')
plt.grid()
```
