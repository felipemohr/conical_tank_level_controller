```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = np.loadtxt("tanque.txt") / 100
```

```python
def getV(h):
  R = 0.33
  H = 0.70
  V = (np.pi / 3) * ((h**3)*(R**2) / (H**2))
  return V
```

```python
f = open('tanque.csv', 'w', newline='')
w = csv.writer(f)

last_v = getV(h[0])

dt = 240.0/h.size
for c in h:
  v = getV(c)
  fo = -(v - last_v)/(dt)
  w.writerow([c, fo])
  last_v = v
```

```python
data = np.genfromtxt('tanque.csv', delimiter=',')
H = data[:,0]
Fo = data[:,1]
```

```python
plt.figure(figsize=(20,10))
plt.plot(H, Fo)
```
